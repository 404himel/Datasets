```python

data_frames = []

for i in range(1, 429):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}')
    dataf = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
    data_frames.append(dataf)

# Concatenate all data frames in the list into one final data frame
dataset = pd.concat(data_frames, ignore_index=True)
```
